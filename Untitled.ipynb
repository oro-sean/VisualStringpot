```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

gate = pd.read_csv('export.csv', index_col=0)
gate['timeStamp'] = pd.to_datetime(gate['timeStamp'], format = "%Y:%m:%d %H:%M:%S")
gate.sort_values(['timeStamp'], inplace=True)
gate['measurement'] = gate.measurement.astype('float64')
gate = gate[gate.measurement > -5]
gate = gate[gate.measurement < 83]
gate['measurement'] = gate.measurement+abs(min(gate.measurement))
gate['measurement'] = gate.measurement/((max(gate.measurement)-min(gate.measurement))/78)
```

```python
loadcells = pd.read_csv('Cyclops.csv', index_col=False)
loadcells['Timestamp'] = pd.to_datetime(loadcells.Timestamp, format="%Y-%m-%d %H:%M:%S").dt.round('1s')
loadcells.rename(columns={'Timestamp': 'timeStamp', 'Backstay ': 'Backstay'}, inplace=True)
loadcells = loadcells.merge(gate,on='timeStamp')
loadcells.rename(columns={'measurement': 'mastGate'}, inplace=True)
loadcells['Mainsheet'] = 2*loadcells.Mainsheet
loadcells['Backstay'] = 2*loadcells.Backstay
```

```python
plotData = loadcells[["timeStamp", "Backstay", "Mainsheet", "mastGate", "Forestay"]]
```

```python
sns.lineplot(x = timeStamp, y = Backstay, data = plotData)
sns.lineplot(x=loadcells.timeStamp,y=loadcells.Forestay)
sns.lineplot(x=loadcells.timeStamp,y=loadcells.Mainsheet)
sns.lineplot(x=loadcells.timeStamp,y=loadcells['Backstay '])
```

```python
time_seriesVars = ["Backstay", "Mainsheet", "mastGate", "Forestay"] # variables to create tiem series of
for var in time_seriesVars: # loop over variables
    plt.figure(figsize=(20,2)) # make elongated figure
    timeSeries = sns.lineplot(data=plotData, x='timeStamp', y=var, lw=0.5) # create time serries plot
    timeSeries.set_title(var+' Vs time') # add title
    sns.despine() # remove outline box
    plt.savefig(var+'_timeSeries.png') # save as .png
    plt.clf() # close plot figure so it is not over plotted
```

```python
plt.figure(figsize=(20,2)) # make elongated figure
timeSeries = sns.lineplot(data=plotData[(plotData.Forestay > 100) & (plotData.Backstay > 1) & (plotData.Mainsheet > 1)], x='timeStamp', y='Backstay', lw=0.5) # create time serries plot
timeSeries = sns.lineplot(data=plotData[(plotData.Forestay > 100) & (plotData.Backstay > 1) & (plotData.Mainsheet > 1)], x='timeStamp', y='Forestay', lw=0.5)
timeSeries = sns.lineplot(data=plotData[(plotData.Forestay > 100) & (plotData.Backstay > 1) & (plotData.Mainsheet > 1)], x='timeStamp', y='Mainsheet', lw=0.5)
timeSeries = sns.lineplot(data=plotData[(plotData.Forestay > 100) & (plotData.Backstay > 1) & (plotData.Mainsheet > 1)], x='timeStamp', y='mastGate', lw=0.5)
timeSeries.set_title('Loads'+' Vs time') # add title
sns.despine() # remove outline box
```
